```python
# import library and module

import tensorflow as tf
from tensorflow import keras

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#load data from google drive
df = pd.read_csv("/content/drive/MyDrive/data/PotatoPrice.csv")
df
```

```python
#Visualization data in graphical representation
plt.xlabel('Potato in kilogram(kg)')
plt.ylabel('price in Taka')
plt.scatter(df.potato_kg, df.price)

plt.show()
```

```python
X = df[['potato_kg']] #since x have to be two dimentional or 2D array. So [[]]
# X
y = df['price']
```

```python
# Split Dataset for Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5)
# X_test
```

```python
# ANN Model Building
model = tf.keras.Sequential([
   tf.keras.layers.Dense(1, input_shape=[1], activation='linear')
])

model.summary()
```

```python
# Compile The ANN model
model.compile(optimizer='sgd',
              loss='mean_squared_error',
             metrics=['accuracy'])
```

```python
#Train ANN Model
model.fit(X_train, y_train, epochs=500)
```

```python
#Predict
model.predict(X_test)
```

```python
#Evaluate The Model Loss and Accuracy
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
test_acc
```

```python
#Predict The Output Manually
print(model.predict([1000]))
```
