```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# reading the image
image = io.imread('./data/ars.jpg')


#io.imshow(image);

# preprocessing
rows, cols = image.shape[0], image.shape[1]
image = image.reshape(rows * cols, 3)

# modelling
kMeans = KMeans(n_clusters = 16)
kMeans.fit(image)

# centers
centers = np.asarray(kMeans.cluster_centers_, dtype = np.uint8)


# labels 
labels = np.asarray(kMeans.labels_, dtype = np.uint8)

labels = np.reshape(labels, (rows, cols))



# reconstructing
newImage = np.zeros((rows, cols, 3), dtype = np.uint8)
for i in range(rows):
    for j in range(cols):
            # assinging every pixel the rgb color of their label's center 
            newImage[i, j, :] = centers[labels[i, j], :]
# io.imsave('dog-compressed.png', newImage); save image 

io.imshow(newImage)
# io.imshow(image)


print(f"Old Image Size : {image.size}")
print(f"Compressed Image Size : {newImage.size}")
```
