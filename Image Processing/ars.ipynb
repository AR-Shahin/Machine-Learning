```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def show_pic(img,title = "Image"): 
    plt.imshow(img, vmin=0, vmax=255)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image1 = cv2.imread('img1.jpg')
image2 = cv2.imread('img2.jpg')
res = image1 + image2
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
show_pic(image1,"image 1")
show_pic(image2,"image 2")
show_pic(weightedSum,"Addition Operation")
```

```python
sub1 = cv2.imread('sub1.jpg')
sub2 = cv2.imread('sub2.jpg')
res = cv2.subtract(sub1, sub2)
show_pic(sub1,"Image 1")
show_pic(sub2,"Image 2")
show_pic(res,"Subtraction Operation")
```

```python
div1 = cv2.imread('div1.png')
div2 = cv2.imread('div2.png')

div1 = cv2.resize(div1, dsize=(240, 240), interpolation=cv2.INTER_CUBIC)
div2 = cv2.resize(div2, dsize=(240, 240), interpolation=cv2.INTER_CUBIC)

res = div1 / div2

show_pic(div1,"Image 1")
show_pic(div2,"Image 2")
show_pic(res,"Division Operation")

```
